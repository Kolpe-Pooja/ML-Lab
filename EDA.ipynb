```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "Consumption of meat per capita.csv"
df = pd.read_csv(file_path)

df_numeric = df.select_dtypes(include=['float64', 'int64']).dropna()

plt.figure(figsize=(10, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', linewidths=0.5)

plt.title("Heatmap of Meat Consumption Correlations")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "Consumption of meat per capita.csv"
df = pd.read_csv(file_path)

Beef_consumption = df["Beef"].dropna()

plt.figure(figsize=(8, 5))
plt.hist(Beef_consumption, bins=10, edgecolor='black', alpha=0.7)

plt.xlabel('Beef Consumption per Capita (kg)')
plt.ylabel('Frequency')
plt.title('Histogram of Beef Consumption per Capita')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "Consumption of meat per capita.csv"
df = pd.read_csv(file_path)

df_avg = df.groupby("Entity")[["Poultry", "Beef", "Sheep and goat", "Pork", "Fish and seafood"]].mean().dropna()

top_countries = df_avg.nlargest(10, "Fish and seafood")

plt.figure(figsize=(10, 5))
top_countries["Fish and seafood"].plot(kind='bar', color='royalblue', edgecolor='black', alpha=0.7)

plt.xlabel("Country")
plt.ylabel("Average Fish Consumption per Capita (kg)")
plt.title("Top 10 Countries with Highest Fish Consumption per Capita")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "Consumption of meat per capita.csv"
df = pd.read_csv(file_path)

df_yearly = df.groupby("Year")["Sheep and goat"].mean().dropna()

plt.figure(figsize=(10, 5))
plt.plot(df_yearly.index, df_yearly.values, marker='o', linestyle='-', color='green', alpha=0.7)

plt.xlabel("Year")
plt.ylabel("Average Sheep and Goat Consumption per Capita (kg)")
plt.title("Trend of Sheep and Goat Consumption Over the Years")
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "Consumption of meat per capita.csv"
df = pd.read_csv(file_path)

plt.figure(figsize=(8, 5))


plt.scatter(df["Year"], df["Beef"], color='red', alpha=0.6, edgecolor='black', label="Beef Consumption")

plt.scatter(df["Year"], df["Poultry"], color='blue', alpha=0.6, edgecolor='black', label="Poultry Consumption")

plt.xlabel("Year")
plt.ylabel("Consumption per Capita (kg)")
plt.title("Scatter Plot of Beef and Poultry Consumption Over Years")

plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "Consumption of meat per capita.csv"
df = pd.read_csv(file_path)

df_avg = df[["Poultry", "Beef", "Sheep and goat", "Pork", "Fish and seafood"]].mean().dropna()

plt.figure(figsize=(8, 8))
plt.pie(df_avg, labels=df_avg.index, autopct='%1.1f%%', colors=['gold', 'red', 'green', 'blue', 'purple'], startangle=140)

plt.title("Average Meat Consumption Distribution")
plt.show()
```
